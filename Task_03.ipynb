```python
import pandas as pd
data=pd.read_csv('student_feedback.csv')
data.head()
```

```python
data.tail(10)
```

```python
data.columns
```

```python
data=data.drop(columns='Unnamed: 0',errors='ignore')
data.columns=[col.strip().replace(' ','_').lower() for col in data.columns]
data
```

```python
data.info()
data.describe(include='all')
```

```python
#Creating barplot
import matplotlib.pyplot as plt
import seaborn as sns

df=data.drop(columns=['student_id'])
avg_ratings=df.mean()

plt.figure(figsize=(10,5))
sns.barplot(y=avg_ratings.index,x=avg_ratings.values)
plt.xlabel('Average Rating')
plt.ylabel('Feedback Category')
plt.title('Average Feedback Ratings')
plt.show()
```

```python
#Creating Heatmap
plt.figure(figsize=(10,5))
sns.heatmap(df.corr(),annot=True)
plt.title('Distribution of Ratings Across Feedback Categories')
plt.xlabel('Ratings')
plt.show()
```

```python
#Creating boxplot
plt.figure(figsize=(10,5))
sns.boxplot(data=df, orient='h', palette='Set1')
plt.title('Distribution of Ratings Across Feedback Categories')
plt.xlabel('Ratings')
plt.show()
```

```python
#Creating histplot
plt.figure(figsize=(10,5))
sns.histplot(data=df,x='degree_of_difficulty_of_assignments',bins=10)
plt.title('Distribution of Overall Student Ratings')
plt.xlabel('Average Ratings')
plt.ylabel('No.of Students')
plt.show()
```

```python
#Creating density plots
plt.figure(figsize=(10,5))
sns.kdeplot(data=df['course_recommendation_based_on_relevance'],fill=True)
plt.title('Distribution of Overall Student Ratings')
plt.xlabel('Average Ratings')
plt.ylabel('No.of Students')
plt.show()
```
